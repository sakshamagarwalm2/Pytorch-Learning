##Reshaping, stacking , squeezing, unsqueeze, permute

```python
x = torch.arange(1.,10.)
x, x.shape
```

```python
# Reshape
x_reshape = x.reshape(3,3)
x_reshape, x_reshape.shape
```

```python
x_reshape = x.reshape(9,1)
x_reshape, x_reshape.shape
```

```python
x_reshape = x.reshape(-1,1)
x_reshape, x_reshape.shape
```

```python
x_reshape = x.reshape(1,-1)
x_reshape, x_reshape.shape
```

```python
#  Stock
z = torch.stack([x,x,x,x],dim=0) #row wise stack
z
```

```python
z = torch.stack([x,x,x,x],dim=1) # col wise stack
z
```

```python
#  Unsqueeze
z = x.unsqueeze(dim=0)
z, z.shape, x, x.shape
```

```python
z = x.unsqueeze(dim=1)
z, z.shape, x, x.shape
```

```python
#  Squeeze
z = z.squeeze()
z, z.shape, x, x.shape
```

```python
# Permute
a = torch.rand(2,3,5,5,6,4,8)
a
```

```python
a.shape
```

```python
a_permuted = a.permute(6,5,4,3,2,1,0)
a_permuted, a_permuted.shape
```

```python
a.shape, a_permuted.shape
```
